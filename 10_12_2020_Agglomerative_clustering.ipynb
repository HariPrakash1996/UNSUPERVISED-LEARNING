```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
air=pd.read_csv(r"C:\Users\sivak\OneDrive\data\trainings\datasets\AirPassengers.csv")
```

```python
air.head()
```

```python
air1=air.select_dtypes(include=[np.number]).copy()
```

```python
air1.head()
```

```python
air1.drop(["Week_num"],axis=1,inplace=True)
```

```python
air1.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
mn=MinMaxScaler()
air2=pd.DataFrame(mn.fit_transform(air1),columns=air1.columns)
```

```python
air2.head()
```

```python
from sklearn.cluster import AgglomerativeClustering
ag=AgglomerativeClustering()
ag.fit(air2)
ag.get_params()
```

```python
ag.compute_full_tree
```

```python
ag.labels_
```

```python
air2["clusters"]=ag.labels_
```
